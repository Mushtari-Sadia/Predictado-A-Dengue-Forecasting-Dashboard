```python
cities = df['City'].unique().tolist()
```

```python
city_dfs=[]
for city in cities:
  city_df = df.groupby('City').get_group(city)
  city_df = city_df.loc[city_df.index >= '2020-02-01']
  city_df = city_df.loc[city_df.index < '2020-02-07']
  city_dfs.append(city_df)
test_df = pd.concat(city_dfs,axis=0)

test_df.to_csv('/content/gdrive/MyDrive/Dengue2/MVH22 DATASET/test.csv')
```
